```python
# Attributes :
# namajurusan (string) : Name of the Major
# dayatampung2019 (int) : Quota for 2019 admissions
# tabel_sebaran_peminat : A HTML Table consists of applicants data
# Universitas : Name of the University the major belongs to
# Jenis : Academic diciplines of each major; saintek (STEM) or soshum (Social Sciences)
```

```python
filename = 'C:/Users/AI LAB/Downloads/clean_data_sbmptn.csv'
data_sbmptn = pd.read_csv(filename)
#jenis=0 saintek, jenis=1 soshum

jenis = []
for i in range(0,len(data_sbmptn)):
    if 'jenis=0' in data_sbmptn.iloc[i,3]:
        jenis.append('saintek')
    elif 'jenis=1' in data_sbmptn.iloc[i,3]:
        jenis.append('soshum')
data_sbmptn['jenis'] = jenis
data_sbmptn = data_sbmptn.iloc[:,4:]
data_sbmptn
```

```python
years = ['2016','2017','2018']

#Extracting the number of applicants from each province in Indonesia.
html_to_datafr(data_sbmptn.iloc[2,2])['2016']
lokasi = ['Aceh','Sumatera Utara','Sumatera Barat','Riau','Jambi','Sumatera Selatan','Bengkulu','Lampung','Kepulauan Bangka Belitung','Kepulauan Riau','DKI Jakarta','Jawa Barat','Jawa Tengah','DI Yogyakarta','Jawa Timur','Banten','Kalimantan Barat','Kalimantan Tengah','Kalimantan Selatan','Kalimantan Timur','Kalimantan Utara','Sulawesi Utara','Sulawesi Tengah','Sulawesi Selatan','Sulawesi Tenggara','Gorontalo','Maluku','Maluku Utara','Papua','Papua Barat','Bali','Nusa Tenggara Barat','Nusa Tenggara Timur']
columns = ['Jurusan','Universitas','Jenis','Jumlah Peminat','Daya Tampung','Nama Jurusan']
columns.extend(lokasi)
data_2016 = pd.DataFrame(index=None, columns=lokasi)
data_2017 = pd.DataFrame(index=None, columns=lokasi)
data_2018 = pd.DataFrame(index=None, columns=lokasi)

for i in range(0,len(data_sbmptn)):
    a = html_to_datafr(data_sbmptn.iloc[i,2])
    clear_output(wait=True)
    print('progress = ',i+1,'/',len(data_sbmptn))
    for loc in lokasi:
        #If there was no applicant, set the number of applicants into zero.
        if loc not in a.index:
            a.at[loc,'2016'] = 0
            a.at[loc,'2017'] = 0
            a.at[loc,'2018'] = 0
            
        data_2016.at[i,loc] = a['2016'][loc]
        data_2017.at[i,loc] = a['2017'][loc]
        data_2018.at[i,loc] = a['2018'][loc]
        
        
```

```python
#Fill the Nan with zeros
data_2016 = data_2016.fillna(0)
data_2017 = data_2017.fillna(0)
data_2018 = data_2018.fillna(0)
```

```python
# We've obtained the number of applicant for each year from every province in Indonesia and the names of the majors.
# The number of applicant for each year are stored in data_2016, data_2017 and data_2018
# While the majors' name stored in data_sbmptn.
# In the next step, we're going to merge the applicants data and the majors data using the ID in each records.
```

```python
#data_sbmptn.join(data_2016, how='outer')
yea = ['2016', '2017', '2018']
data_years = [ data_2016, data_2017, data_2018]

#assign each major with an ID
data_sbmptn['id_jurusan'] = [i for i in range(0,len(data_sbmptn))]

#Iterate through the applicants data
for year,data_each_year in zip(years,data_years):
    
    flattened= pd.DataFrame()
    flattened['Pendaftar'] = data_each_year.values.flatten().astype(int).tolist()
    jur_ids = []
    loc_ids = []
    for major in range(0,len(data_sbmptn)):
        for prov in lokasi:
            loc_ids.append(prov)
            jur_ids.append(major)
    flattened['Asal Daerah'] = loc_ids
    flattened['id_jurusan'] = jur_ids
    flattened['Tahun'] = [int(year) for k in range(0,len(loc_ids))]
    pd.merge(data_sbmptn, flattened, on='id_jurusan').drop(['tabel_sebaran_peminat','dayatampung2019'], axis=1).to_excel('data_'+year+'_flatten.xlsx')
    #We've dropped the html table since we thought it will burden the size of the data, and we already extracted the necessary.
    #We've also dropped the dayatampung2019 since we'll not going to use the column for our analysis
    #We saved the dataframe as excel.
```

```python
data_2016_f = pd.read_excel('data_2016_flatten.xlsx', sheetname='Sheet1')
data_2017_f = pd.read_excel('data_2017_flatten.xlsx', sheetname='Sheet1')
data_2018_f = pd.read_excel('data_2018_flatten.xlsx', sheetname='Sheet1')
```
